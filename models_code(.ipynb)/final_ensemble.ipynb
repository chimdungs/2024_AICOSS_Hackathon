```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
```

```python
model_1 = pd.read_csv("./model_1.csv") #efficietnet_v2_s_no_val

model_2 = pd.read_csv("./model_2.csv") #efficientb4_no_validation

model_3 = pd.read_csv("./model_3.csv") #0.951

model_4 = pd.read_csv("./model_4.csv") #0.949
```

```python
def sigmoid(x): #사용자 정의 시그모이드 함수(후에 csv파일에 적용하기 위해)
    return 1 / (1 + np.exp(-x))
```

```python
# 1열의 string(문자열)부분을 제거한 부분을 따로 변수명으로 저장합니다.
model_1_csv = model_1.iloc[:, 1:]

model_2_csv = model_2.iloc[:,1:]

model_3_csv = model_3.iloc[:,1:]

model_4_csv = model_4.iloc[:,1:]
```

## 앙상블
- 여러 분류기의 결과(logit 점수)상태에서 적절하게 가중치를 주어 soft voting의 방법으로 앙상블(ensemble)해줍니다.
- 각각의 가중치 = 40%, 30%, 20%, 10%로 배분

```python
submit = pd.read_csv('./sample_submission.csv')

submit.iloc[:,1:] = (0.4*model_1_csv + 0.3*model_2_csv + 0.2*model_3_csv + 0.1*model_4_csv).apply(sigmoid)
submit.head() # final_submission
```

```python
submit.to_csv('./final_submission.csv', index=False)
```
